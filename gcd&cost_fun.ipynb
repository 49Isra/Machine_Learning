## Gradient Descent Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13]) 

m = 0
b = 0

learning_rate = 0.01
epochs = 1000
n = len(x) 

for i in range(epochs):

    y_pred = m * x + b
    
    dm = (-2/n) * sum(x * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db 

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("test_scores.csv")

x = data["math"].values
y = data["cs"].values

m = 0
b = 0

learning_rate = 0.00001
epochs = 100000

n = len(x)

for i in range(epochs):

    y_pred = m * x + b

    dm = (-2/n) * sum(x * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db


print("Slope (m):", m)
print("Intercept (b):", b)


math_mark = 75
predicted_cs = m * math_mark + b

print("Predicted CS:", predicted_cs)
```
